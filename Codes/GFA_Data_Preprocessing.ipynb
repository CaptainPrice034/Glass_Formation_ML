```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Metallic_Glass_Forming_with_features.csv')
```

```python
df.head()
```

```python
def classify_glass_form(y):
    if y >= 0.6:
        return 'Good'
    elif y <= 0.56:
        return 'Poor'
    else:
        return 'Moderate'

df['Glass form ability'] = df['Trg'].apply(classify_glass_form)
```

```python
sns.countplot(x='Glass form ability', data=df, palette=['red','yellow','green'])
plt.xlabel('Glass Form Ability')
plt.ylabel('Count')
plt.title('Distribution of Glass Form Ability')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Trg'], bins=5, kde=True, color='skyblue')
plt.xlabel('Trg')
plt.ylabel('Frequency')
plt.title('Histogram of Trg')
```

```python
df_corr = df.drop(['Material compositions', 'main_element', 'Glass form ability'], axis=1)
```

```python
corr_matrix = df_corr.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    annot=False,
    cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'}
)

plt.title('Correlation Matrix', fontsize=14, pad=15)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Save the plot
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')  # PNG format, high resolution

plt.show()
```

```python
upper_tri = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)
```

```python
high_corr_pairs = upper_tri.stack()[
    (upper_tri.stack() > 0.95) | (upper_tri.stack() < -0.95)
]
```

```python
print("Highly correlated feature pairs (|corr| > 0.95):")
print(high_corr_pairs.sort_values(ascending=False))
```
